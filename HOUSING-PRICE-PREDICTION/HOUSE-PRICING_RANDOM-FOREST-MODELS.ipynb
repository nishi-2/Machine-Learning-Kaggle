#### The main aim is to get the best random forest model

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
file_path = 'C:/Users/hp/Desktop/MACHINE LEARNING/HOUSING-PRICE-PREDICTION/train.csv'
home_data = pd.read_csv(file_path)
y = home_data.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = home_data[features]
train_X, test_X, train_y, test_y = train_test_split(X,y, random_state=0)
```

Now that out data is ready to use, we will use it in five different models and see the outcome from each.

```python
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='mae', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth = 7, random_state=0)
```

To select the best model out of the five, we define a function score_model() below. This function returns the mean absolute error (MAE) from the validation set. Recall that the best model will obtain the lowest MAE.

```python
def score_model(model, Xtrain=train_X, Xtest=test_X, ytrain=train_y, ytest=test_y):
    model.fit(Xtrain,ytrain)
    pred = model.predict(Xtest)
    return mean_absolute_error(ytest,pred)

models = [model_1,model_2,model_3,model_4,model_5]
```

```python
for i in range(0,len(models)):
    mae = score_model(models[i])
    print("Model %d is having MAE as %d "%(i+1,mae))
```

##### So, as per the above data, we can see that the best model is the model_5 having the minimum mean absolute error

Now, the prediction generated can be saved into a file and stored.

```python
best_model = model_5
best_model.fit(train_X,train_y)
prediction = best_model.predict(test_X)
```

```python
output = pd.DataFrame({
    'ID':test_X.index,
    'SalePrice':prediction
})
output.to_csv('predicted-result.csv',index=False)
```
